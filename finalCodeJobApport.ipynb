```python
import pandas as pd
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import pycountry

# Define a function to clean the job titles
def clean_job_title(job_title):
    job_title_lower = job_title.lower()
    job_title_tokens = word_tokenize(job_title_lower)
    stop_words = set(stopwords.words('english'))
    job_title_filtered = [word for word in job_title_tokens if word not in stop_words]
    lemmatizer = WordNetLemmatizer()
    job_title_lemmatized = [lemmatizer.lemmatize(word) for word in job_title_filtered]
    special_chars_pattern = r'[^a-zA-Z\s]'
    job_title_cleaned = re.sub(special_chars_pattern, '', ' '.join(job_title_lemmatized)).strip()
    return job_title_cleaned

# Get a dictionary of country names and their ISO codes using the pycountry library
country_names = {country.name.lower(): country.name for country in pycountry.countries}

# Add more common country names if needed
additional_country_names = {'united states': 'United States', 'united kingdom': 'United Kingdom', 'united arab emirates': 'United Arab Emirates', 'singapore': 'Singapore', 'india': 'India', 'brazil': 'Brazil', 'italy': 'Italy', 'spain': 'Spain', 'sweden': 'Sweden', 'norway': 'Norway', 'denmark': 'Denmark'}
country_names.update(additional_country_names)

# Define a function to extract only the country name from the "Country" column
def extract_country(country_text):
    # Convert to lowercase for case-insensitive comparison
    country_text_lower = country_text.lower()

    # Check if the entire text is a valid country name
    if country_text_lower in country_names:
        return country_names[country_text_lower]

    # If not, check for individual words that match country names
    country_words = country_text_lower.split()
    valid_country_words = [word for word in country_words if word in country_names]
    if valid_country_words:
        return country_names[valid_country_words[-1]]

    # If no valid country names found, return the original text
    return country_text

# Define a function to clean the job titles and extract country and remote information
def clean_and_separate_job_title(job_title, location):
    job_title_cleaned = clean_job_title(job_title)

    # Check for country names and "remote" keyword
    remote_keyword = 'remote'

    countries_in_title = [country_names[word] for word in country_names.keys() if word in job_title_cleaned]
    country_from_location = location.strip()

    if countries_in_title:
        country_from_title = countries_in_title[0]
        if country_from_title.lower() == country_from_location.lower():
            country = country_from_title
            country_from_location = ''
        else:
            country = country_from_location
            country_from_title = ''
    else:
        country = country_from_location
        country_from_title = ''

    job_title_cleaned = job_title_cleaned.replace(country, '').replace(country_from_title, '').strip()

    is_remote = int(remote_keyword.lower() in location.lower() or remote_keyword.lower() in job_title_cleaned)
    job_title_cleaned = job_title_cleaned.replace(remote_keyword, '').strip()

    return job_title_cleaned, country, is_remote

# Your CSV data as a DataFrame


df = dataFrame
# Clean and separate the "Job Title" column using the defined function
df["Job Title"], df["Country"], df["Is Remote"] = zip(*df.apply(lambda row: clean_and_separate_job_title(row["Job Title"], row["Location"]), axis=1))

# Extract only the country name from the "Country" column
df["Country"] = df["Country"].apply(extract_country)

print(df)
```

```python
def split_comma_column_values (colum_name_comma_value,dataFrame) :
    dataFrame[colum_name_comma_value] = dataFrame[colum_name_comma_value].str.split(',')
    new_df = dataFrame.explode(colum_name_comma_value)
    new_df = new_df.reset_index(drop=True)
    new_df[colum_name_comma_value].str.strip()
    new_df.drop_duplicates(inplace=True)
    return new_df
```
